```python
# Define a class called Library
class Library():
  def __init__(self, name, location):
    self.books = []
    self.lib_name = name
    self.lib_location = location

  def add_book(self, new_book_name):
    self.books.append(new_book_name)
    return self.books

  def remove_book(self, remove_book):
    if remove_book in self.books:
      self.books.remove(remove_book)
    else:
      print(f"'{remove_book}' not found in the library.")

  def get_book(self):
    return self.books

# Define a class called Book
class Book():
  def __init__(self, name, author, availability):
    self.Book_name = name
    self.Author_name = author
    self.Availability = availability

  def borrow(self):
    if self.Availability:
      self.Availability = False
      print(f"You borrowed '{self.Book_name}'")
    else:
      print(f"'{self.Book_name}' is not available")

  def return_book(self):
    self.Availability = True
    print(f"'{self.Book_name}' has been returned")

# Create Library instance
Mylib = Library("The Library", "Karachi")
print("Library Name:", Mylib.lib_name)
print("Location:", Mylib.lib_location)
print("Books in library:", Mylib.get_book())

# Create Book instances
book1 = Book("Poetry", "Allama Iqbal", True)
book2 = Book("Comedy", "Umer Sharif", True)
book3 = Book("Tragedy", "Ahmed", True)

# Add books to library
Mylib.add_book(book1.Book_name)
Mylib.add_book(book2.Book_name)
Mylib.add_book(book3.Book_name)

# Show updated book list
print("Books after adding:", Mylib.get_book())

# Borrow a book
book1.borrow()
print("Is 'Poetry' available now?", book1.Availability)

# Remove a book
Mylib.remove_book("Tragedy")
print("Books after removing 'Tragedy':", Mylib.get_book())
```

```python
# üìö Book class represents each individual book
class Book:
    def __init__(self, name, author, available=True):
        self.name = name               # Book title
        self.author = author           # Author of the book
        self.available = available     # Is the book available to borrow? (True by default)

    # üèÉ Method to borrow the book
    def borrow(self):
        if self.available:
            self.available = False     # Book is now borrowed (not available)
            print(f"You borrowed '{self.name}' by {self.author}")
        else:
            print(f"'{self.name}' is not available right now.")  # Already borrowed

    # üîÅ Method to return the book
    def return_book(self):
        self.available = True          # Book is now available again
        print(f"'{self.name}' has been returned.")

# üèõÔ∏è Library class represents a library that stores books
class Library:
    def __init__(self, name, location):
        self.name = name               # Library name
        self.location = location       # Library location
        self.books = []                # List to store Book objects

    # ‚ûï Method to add a book to the library
    def add_book(self, book):
        self.books.append(book)        # Add Book object to list
        print(f"'{book.name}' added to the library.")

    # ‚ûñ Method to remove a book from the library by title
    def remove_book(self, book_name):
        for book in self.books:        # Go through each book
            if book.name == book_name: # If the book title matches
                self.books.remove(book)
                print(f"'{book_name}' removed from the library.")
                return
        print(f"'{book_name}' not found in the library.")  # If not found

    # üëÄ Method to display all books with their status
    def show_books(self):
        if not self.books:
            print("No books in the library.")  # Empty library
        else:
            print("Books in the library:")
            for book in self.books:
                status = "Available" if book.available else "Borrowed"
                print(f"- {book.name} by {book.author} ({status})")

# ‚úÖ Example usage (testing)

# Create a library
library = Library("The Knowledge Hub", "Hyderabad")

# Create some books
book1 = Book("Science", "Dr. Qasim")
book2 = Book("Math", "Prof. Ali")
book3 = Book("History", "Dr. Fatima")

# Add books to the library
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Show all books in the library
library.show_books()

# Borrow a book
book1.borrow()

# Try to borrow it again (should say it's not available)
book1.borrow()

# Show updated list
library.show_books()

# Return the book
book1.return_book()

# Show updated list
library.show_books()

# Remove a book
library.remove_book("Math")

# Show final list of books
library.show_books()
```
